```python
# 2ª QUESTÃO

lista1 = ['Marcelo', 2, 3, 4.2, 5, 6, 'a', 'e', 'i', 'O', 'U']
lista2 = ['MarCelo', 8, 9, 3, 4.2, 1, 'I', 'j', 'e', 'O', 'B']

lista_mod1 = []
lista_mod2 = []

repetidos = []

print(lista1)
print(lista2)
print()

for i in lista1:
    if type(i) == str:        
        i = i.lower()
    
    lista_mod1.append(i)

for i in lista2:
    if type(i) == str:        
        i = i.lower()
    
    lista_mod2.append(i)


print('Valores comuns as duas listas: ')
for i in lista_mod1:
    if i in lista_mod2:
        print(i, end=' ')

print()
print()

print('Os valores que só existem na primeira lista')
for i in lista_mod1:
    if i not in lista_mod2:
        print(i, end=' ')
        repetidos.append(i)

print()
print()

print('Os valores que existem apenas na segunda lista')
for i in lista_mod2:
    if i not in lista_mod1:
        print(i, end=' ')
        repetidos.append(i)

print()
print()

print('Uma lista com os elementos não repetidos das duas listas')
print(repetidos)

```

```python
# Inicializa as variáveis
cidades = []
casos_por_cidade = []
maior_casos_por_cidade = []
dia_maior_casos_por_cidade = []
media_casos_por_cidade = []

continua_inserindo = True

# Loop para inserção de dados
while continua_inserindo:
    cidade = input("Digite o nome da cidade: ")
    data = input("Digite a data da coleta: ")
    casos = int(input("Digite o número de casos de COVID-19: "))

    if cidade in cidades:
        indice_cidade = cidades.index(cidade)
        casos_por_cidade[indice_cidade].append(casos)
    else:
        cidades.append(cidade)
        casos_por_cidade.append([casos])
        maior_casos_por_cidade.append(casos)
        dia_maior_casos_por_cidade.append(data)
        media_casos_por_cidade.append(casos)

    # Pergunta ao usuário se deseja continuar inserindo dados
    resposta = input("Deseja inserir mais dados? (s/n) ")
    if resposta.lower() == "n":
        continua_inserindo = False

# Calcula as informações requeridas
maior_acumulado_cidade = ""
maior_acumulado_valor = 0
for i, cidade in enumerate(cidades):
    acumulado = sum(casos_por_cidade[i])
    if acumulado > maior_acumulado_valor:
        maior_acumulado_cidade = cidade
        maior_acumulado_valor = acumulado

    media = sum(casos_por_cidade[i]) / len(casos_por_cidade[i])
    media_casos_por_cidade[i] = media

    maior_dia = dia_maior_casos_por_cidade[i]
    maior_valor = maior_casos_por_cidade[i]
    for j, casos in enumerate(casos_por_cidade[i]):
        if casos > maior_valor:
            maior_valor = casos
            maior_dia = j+1

    maior_casos_por_cidade[i] = maior_valor
    dia_maior_casos_por_cidade[i] = f"Dia {maior_dia}"

# Imprime as informações
print("Número total de casos de COVID-19 em cada cidade:")
for i, cidade in enumerate(cidades):
    acumulado = sum(casos_por_cidade[i])
    print(f"{cidade}: {acumulado}")

print("Dia com o maior número de casos em cada cidade:")
for i, cidade in enumerate(cidades):
    print(f"{cidade}: {dia_maior_casos_por_cidade[i]} ({maior_casos_por_cidade[i]} casos)")

print("Média de casos diários em cada cidade:")
for i, cidade in enumerate(cidades):
    print(f"{cidade}: {media_casos_por_cidade[i]} casos/dia")

print(f"Cidade com maior número acumulado de casos até o momento: {maior_acumulado_cidade} ({maior_acumulado_valor} casos)")
```

```python
dados_diarios = []

while True:
    cidade = input('Digite a cidade da coleta: ').upper()
    data = input('Digite a data da coleta: ')
    casos = int(input('Digite o número de casos de COVID-19 na cidade: '))

    dados_diarios.append([cidade, data, casos])

    print(f'Dados da coleta: Cidade: {cidade}, Data: {data}, Casos: {casos}')

    c = int(input('Digite 1 para continuar com a coleta ou 0 para encerrar.'))

    if c == 0:
        break

# Criar um dicionário para armazenar a soma total de casos em cada cidade
soma_cidades = {}

# Iterar sobre a lista "dados_diarios" e calcular a soma total de casos em cada cidade
for dados in dados_diarios:
    cidade = dados[0]
    casos = dados[2]
    if cidade in soma_cidades:
        soma_cidades[cidade] += casos
    else:
        soma_cidades[cidade] = casos

# Exibir a soma total de casos em cada cidade
for cidade, soma in soma_cidades.items():
    print(f'{cidade}: {soma}')
```
